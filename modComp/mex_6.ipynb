```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Generate random variables
n = 1000

# Create n Bernoulli random variables
Z = np.random.binomial(n=1, p=0.5, size=n)

# Generate Poisson random variables based on Z
X = np.where(Z == 0, np.random.poisson(lam=1, size=n), np.random.poisson(lam=5, size=n))
Y = np.where(Z == 0, np.random.poisson(lam=1, size=n), np.random.poisson(lam=5, size=n))
```

```python
### Plot the scatter plot of Y against X
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].scatter(X, Y, alpha=0.3)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Scatter plot of Y against X')

axs[1].scatter(X[Z == 0], Y[Z == 0], alpha=0.3)
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
axs[1].set_title('Scatter plot of Y against X when Z is 0')

axs[2].scatter(X[Z == 1], Y[Z == 1], alpha=0.3)
axs[2].set_xlabel('X')
axs[2].set_ylabel('Y')
axs[2].set_title('Scatter plot of Y against X when Z is 1')

plt.tight_layout()
plt.show()
```
